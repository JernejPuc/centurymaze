# Colour palettes

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic (objects)

```python
clrs = np.array([
    [155, 205, 0],
    [0, 145, 0],
    [255, 145, 0],
    [0, 0, 255],
    [0, 160, 200],
    [255, 0, 0],
    [100, 50, 0],
    [115, 20, 165],
    [215, 20, 135]])

_, axes = plt.subplots(3, 3, figsize=(12, 12))

for i in range(3):
    for j in range(3):
        rgb = clrs[i*3 + j]
        sat = round((rgb.max() - rgb.min()) / np.maximum(rgb.max(), 1e-6), 3)

        axes[i, j].imshow(np.ones((1, 1, 3), dtype=np.int32) * rgb[None, None])
        axes[i, j].set_xticks([])
        axes[i, j].set_yticks([])

        axes[i, j].set_title(f'Saturation: {sat:.3f}');
```
